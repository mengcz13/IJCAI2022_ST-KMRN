```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pems_bay = pd.read_hdf('../../data/pems/pems-bay.h5')
```

```python
pems_bay = np.load('../../data/pems/dense_multires/PEMS-BAY/sampled_dense_multires_seqs.npz', allow_pickle=True)
list(pems_bay.keys())
```

```python
pems_bay_values = pems_bay.values
```

```python
pems_bay_values.shape
```

```python
node_i = 2
node_seq = pems_bay_values[:, node_i].reshape(-1, 1)
print(node_seq.shape)

fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(node_seq[:, 0])
ax.set_ylim(0, 80)
ax.set_title('Whole Seq')

fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(node_seq[288*7:288*9, 0])
ax.set_ylim(0, 80)
ax.set_title('Two Days')

fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(node_seq[288*7*3:288*7*4, 0])
ax.set_ylim(0, 80)
ax.set_title('Two Weeks')

fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(node_seq[:288*60, 0])
ax.set_ylim(0, 80)
ax.set_title('Two Months')

fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(node_seq[:288*180, 0])
ax.set_ylim(0, 80)
ax.set_title('Six Months')
```
