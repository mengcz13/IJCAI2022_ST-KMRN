```python
import os
import sys
import re
from datetime import datetime, timedelta

import sparse
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = sparse.load_npz('../../data/solar_energy/volume_valid_region_borough.npz').todense()
data.shape
```

```python
node_ids = np.linspace(0, data.shape[1], num=5, endpoint=False, dtype=np.int64)
horizon = 10 * 48
start_ts = np.linspace(0, data.shape[0] - horizon, num=5, endpoint=False, dtype=np.int64)

fig, axes = plt.subplots(len(node_ids), len(start_ts), figsize=(40, 10))
for ni in range(len(node_ids)):
    for ti in range(len(start_ts)):
        ax = axes[ni][ti]
        ax.plot(data[ti:ti + horizon, ni, :])
        ax.set_title('Node {}, StartT {}'.format(node_ids[ni], start_ts[ti]))
fig.tight_layout()
```
